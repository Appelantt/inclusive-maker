# 01 - Exploration du signal EEG

Ce notebook permet d'explorer des signaux EEG synthétiques ou réels.

## Objectifs
- Visualiser les 8 canaux EEG
- Observer les bandes de fréquence
- Tester les filtres

```python
import numpy as np
import matplotlib.pyplot as plt
from inclousive_maker.signal_processing.features import compute_all_bandpowers
from inclousive_maker.shared.constants import EEG_SAMPLING_RATE, BANDS
```

```python
# Générer un signal synthétique
t = np.arange(EEG_SAMPLING_RATE) / EEG_SAMPLING_RATE
signal = 10 * np.sin(2 * np.pi * 10 * t) + np.random.normal(0, 1, len(t))

plt.figure(figsize=(12, 4))
plt.plot(t, signal)
plt.title('Signal EEG synthétique (10 Hz alpha)')
plt.xlabel('Temps (s)')
plt.ylabel('Amplitude')
plt.show()
```
